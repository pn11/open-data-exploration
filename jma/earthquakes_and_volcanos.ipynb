```python
import feedparser
import json
import pandas as pd
import requests
import time
from tqdm import tqdm
import xml.dom.minidom
import xml.etree.ElementTree as ET

print(feedparser.__version__)
print(pd.__version__)
print(requests.__version__)
```

```python
feed = feedparser.parse("http://www.data.jma.go.jp/developer/xml/feed/eqvol_l.xml")
feed
```

```python
feed['entries']
```

```python
feed_simple = []
for f in feed['entries']:
    feed_simple.append({
        'type': f['title'],
        'updated': f['updated'],
        'updated_parsed': f['updated_parsed'],
        'link': f['links'][0]['href']
    })
print(feed_simple)
```

```python
pd.DataFrame.from_dict(feed_simple)
```

```python
feed_entry1 = feedparser.parse(feed_simple[0]['link'])
feed_entry1
```

```python
feed_entry1['feed']['summary']
```

```python
# 地震情報の最初のエントリーを出力
for fs in feed_simple:
    if fs['type'] == '震源・震度に関する情報':
        f = feedparser.parse(fs['link'])
        print(f)
        xml_ = ET.fromstring('<root>'+f['feed']['summary']+'</root>') # xmlの開始と終了タグを追加
        xmltmp = xml.dom.minidom.parseString(ET.tostring(xml_, encoding='unicode'))
        print(xmltmp.toprettyxml(indent=' ', newl=''))
        break
```

```python
feed_earthquake_hypocenter_intensity = []
for fs in tqdm(feed_simple):
    if fs['type'] == '震源・震度に関する情報':
        f = feedparser.parse(fs['link'])
        xml_ = ET.fromstring('<root>'+f['feed']['summary']+'</root>') # xmlの開始と終了タグを追加
        feed_earthquake_hypocenter_intensity.append({
            'link': fs['link'],
            'hypocenter area name': xml_.find('earthquake/hypocenter/area/name').text,
            'hypocenter area code': xml_.find('earthquake/hypocenter/area/code').text,
            'hypocenter coordinate description': xml_.find('earthquake/hypocenter/area/coordinate').attrib['description'],
            'hypocenter coordinate': xml_.find('earthquake/hypocenter/area/coordinate').text,
            'magnitude': xml_.find('earthquake/magnitude').text,
            'intensity': xml_.find('intensity')
        })
        time.sleep(0.5)
pd.DataFrame.from_dict(feed_earthquake_hypocenter_intensity)
```
